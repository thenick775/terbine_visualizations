```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/new_york_fare_Card/data/newyorkdat_multi_all.csv")
```

```python
#process data in format suitable to pie chart
```

```python
dfpie=df[["station"]].drop_duplicates()
```

```python
dfpie=dfpie.reset_index(drop=True)
```

```python
dfpie["total_percentage"]=np.nan
```

```python
dfpie
```

```python
df.columns
```

```python
for ind,row in dfpie.iterrows():
    rowsum=df[df["station"]==row["station"]][['full_fare',
       'senior_citizen_disabled',
       '_7_day_ada_farecard_access_system_unlimited',
       '_30_day_ada_farecard_access_system_unlimited',
       'joint_rail_road_ticket', '_7_day_unlimited', '_30_day_unlimited',
       '_14_day_reduced_fare_media_unlimited', '_1_day_unlimited',
       '_14_day_unlimited', '_7_day_express_bus_pass',
       'transit_check_metrocard', 'lib_special_senior',
       'rail_road_unlimited_no_trade',
       'transit_check_metrocard_annual_metrocard',
       'mail_and_ride_ezpay_express', 'mail_and_ride_unlimited', 'path_2_trip',
       'airtrain_full_fare', 'airtrain_30_day', 'airtrain_10_trip',
       'airtrain_monthly']].sum(axis = 0, skipna = True).sum(axis = 0, skipna = True)
    print(rowsum)
    dfpie.iloc[ind,1]=rowsum
```

```python
dfpie
```

```python
total_all=dfpie["total_percentage"].sum(axis = 0, skipna = True)
print(total_all)
```

```python
dfpie["total_percentage"]=dfpie["total_percentage"].apply(lambda x: float(x)/total_all)
```

```python
dfpie
```

```python
dfpie.to_csv("/new_york_fare_Card/data/dfpie.csv",index=None)
```
