```python
dfch0["DateTimeUTC"]=dfch0["Date"]+" "+dfch0["TimeUTC"]
dfch1["DateTimeUTC"]=dfch1["Date"]+" "+dfch1["TimeUTC"]
dfch2["DateTimeUTC"]=dfch2["Date"]+" "+dfch2["TimeUTC"]
dfch3["DateTimeUTC"]=dfch3["Date"]+" "+dfch3["TimeUTC"]
```

```python
dfch0=dfch0[["DateTimeUTC","DigitalCounts","LinearMicrostrain"]]
dfch1=dfch1[["DateTimeUTC","DigitalCounts","LinearMicrostrain"]]
dfch2=dfch2[["DateTimeUTC","DigitalCounts","LinearMicrostrain"]]
dfch3=dfch3[["DateTimeUTC","DigitalCounts","LinearMicrostrain"]]
```

```python
alllst=[dfch0,dfch1,dfch2,dfch3]
```

```python
for x,df in enumerate(alllst):
    for i,f in enumerate(df.columns):
        if i>0:
            print(df.columns[i])
            df=df.rename(columns={df.columns[i]:df.columns[i]+"_ch"+str(x)})
    print(df.columns)
    print("\n\n")
    alllst[x]=df
```

```python
result=pd.merge(alllst[0],alllst[1],on="DateTimeUTC")
result=pd.merge(result,alllst[2],on="DateTimeUTC")
result=pd.merge(result,alllst[3],on="DateTimeUTC")
```

```python
result
```

```python
def runs(x):
    if x=="999999":
        return np.nan
    else:
        return x
```

```python
result=result.applymap(runs).dropna().reset_index(drop=True).copy()
```

```python
result=result[['DateTimeUTC', 'LinearMicrostrain_ch0', 'LinearMicrostrain_ch1',
               'LinearMicrostrain_ch2', 'LinearMicrostrain_ch3']]
```
