```python
import pandas as pd
import matplotlib as plt
import matplotlib.cm as cm
import colorsys
```

```python
df=pd.read_csv("/ausphoto.csv")
```

```python
df
```

```python
df["REGIONID"].drop_duplicates()
```

```python
df=df.sort_values('LASTCHANGED')
```

```python
df
```

```python
#data can be edited at source, sort by last changed date for data, and then keep the last
df.drop_duplicates(subset=["INTERVAL_DATETIME","REGIONID"],inplace=True,keep='last')
```

```python
df
```

```python
strip=df[["INTERVAL_DATETIME","REGIONID","POWER"]].copy()
```

```python
strip
```

```python
#reshape dataframe for easy sequential access in d3
pivoted=strip.pivot(index='INTERVAL_DATETIME', columns='REGIONID', values='POWER')
```

```python
pivoted
```

```python
sumcolqld=pivoted["QLD1"]+pivoted["QLDC"]+pivoted["QLDN"]+pivoted["QLDS"]
sumcoltas=pivoted["TAS1"]+pivoted["TASN"]+pivoted["TASS"]

pivoted["QLDA"]=sumcolqld
pivoted["TASA"]=sumcoltas
```

```python
pivoted
```

```python
pivotedtocolor=pivoted[["NSW1","QLDA","SA1","TASA","VIC1"]].copy()
```

```python
pivotedtocolor
```

```python
#generate chart from raw data, binned to 1 day and averaged
pivotedtocolortmp=pivotedtocolor.copy()
pivotedtocolortmp.index = pd.to_datetime(pivotedtocolortmp.index)
p=pivotedtocolortmp.groupby(pd.Grouper(freq='1D')).mean().plot(figsize=(45,20))
hours = plt.dates.HourLocator(interval = 72)

p.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=True)

p.xaxis.set_major_locator(hours)
p.xaxis.set_major_formatter(plt.dates.DateFormatter('%m/%d %H:%M:%S'))
p.tick_params(axis='x', labelrotation=45)

p.set_ylabel("Generation (MW)",fontsize = 18)
p.set_xlabel("Date",fontsize = 18)
p.set_title('Solar Generation Average',fontsize= 30)

p.legend(prop={'size': 30}) #manually set legend
```

```python
pivotedtocolor.to_csv("/dfraw1.csv")
```

```python
#map dataframe values to color range provided by mapper
minimadf = min(df["POWER"])
maximadf = max(df["POWER"])
print(minimadf)
print(maximadf)

norm = plt.colors.Normalize(vmin=minimadf, vmax=maximadf, clip=True)
mapper = cm.ScalarMappable(norm=norm, cmap=cm.get_cmap("plasma"))

def tocolor(x):
    t=list(mapper.to_rgba(x))
    return colorsys.rgb_to_hls(t[0], t[1], t[2])[0]

pivotedtocolor[["NSW1","QLDA","SA1","TASA","VIC1"]]=pivotedtocolor[["NSW1","QLDA","SA1","TASA","VIC1"]].applymap(tocolor)
```
